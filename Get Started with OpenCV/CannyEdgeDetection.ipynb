```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('Images/bike.jpg')
```

```python
img
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
resized_img=cv2.resize(img,(int(img.shape[1]*2),int(img.shape[0]*2)))
cv2.imshow('Bike',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Edge detection
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur=cv2.GaussianBlur(gray,(5,5),0)
canny=cv2.Canny(blur,threshold1=200,threshold2=210)
```

```python
resized_img=cv2.resize(canny,(int(img.shape[1]*2),int(img.shape[0]*2)))
cv2.imshow('Bike',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Export edges
cv2.imwrite('Images/edges.jpg',canny)
```
