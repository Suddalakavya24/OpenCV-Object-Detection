```python
def take_photo():
    cap=cv2.VideoCapture(0)
    ret,frame=cap.read()
    cv2.imwrite('Images/webcamPic.jpg',frame)
    cap.release()
    
```

```python
take_photo()
```

```python
#Rendering in RealTime
cap=cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame=cap.read()
    cv2.imshow('Webcam',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
