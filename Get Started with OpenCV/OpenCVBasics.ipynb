```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Images//bike.jpg", cv2.IMREAD_COLOR)

cv2.imshow('Image',img)# BGR
plt.imshow(img)#RGB

print(img)
print(img.shape)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To convert BGR TO RGB

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Images//bike.jpg", cv2.IMREAD_COLOR)
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(RGB_img)
```

```python
# Open image in grey scale mode

import cv2
import matplotlib.pyplot as plt

# Using 0 to read image in grayscale mode
img = cv2.imread('Images//bike.jpg', 0)
rgbimage=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(rgbimage)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Storing Image
import cv2
import os

directory = r'C://Users//sudda//Desktop//OpenCV-Object Detection//Get Started with OpenCV//Images'
img = cv2.imread('Images//bike.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
os.chdir(directory)

print("Before saving image:")
print(os.listdir(directory))

newfilename = 'EditedBike.jpg'
cv2.imwrite(newfilename, img)


print("After saving image:")
print(os.listdir(directory))
print('Successfully saved')
```
